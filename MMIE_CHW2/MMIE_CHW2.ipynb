## 2. KMeans Clustering (LSE approach)

You are given a dataset of **600** points in a **2-dimensional** space, which have come from **3 different distributions**. The dataset is stored in a file called **Image.csv**, which has two columns: **x** and **y**. You want to use **K-means** to cluster the data points into **K** groups, and analyze the results.

- Part 1: Write a Python function called **load_data** that takes the file name **Image.csv** as an input and returns a numpy array **X**. **X** should be a **600 by 2** matrix, where each row is a point in the 2-dimensional space. ``(5 pt.)``

````python
def load_data(file_name):
    X=np.zeros((600,2))
    file = open(file_name)
    csvreader = csv.reader(file)
    i=0
    for row in csvreader:
        X[i]=np.array(row)
        i+=1
    return X
````

- Part 2: Write a Python function called **kmeans** that takes **X** and **K** as inputs and returns two numpy arrays: **M** and **U**. Use the **K-means(LSE approach)** algorithm as described below to find the optimal cluster centroids and assignments. ``(15 pt.)``

The general approach of the k-means clustering algorithm is as follows:

- The algorithm takes two inputs: the data matrix X of size n by d, where n is the number of data points and d is the dimension of each data point, and the number of clusters K.
- The algorithm outputs two matrices: the cluster centroid matrix M of size K by d, where each row represents the coordinates of a cluster center, and the cluster assignment matrix U of size n by K, where each row represents the membership of a data point to a cluster.
- The algorithm works by first initializing the M matrix randomly from the data points in X; and then iteratively updating the cluster centroids and the cluster assignments until convergence, which means that the total squared error between the data points and their assigned cluster centroids does not change.
- The algorithm uses the least squares error (LSE) method to measure the quality of the clustering. The LSE method is a way of quantifying how well the cluster centroids fit the data points, by minimizing the sum of the squared distances between each data point and its assigned cluster centroid. The LSE method is used in both steps of the algorithm: assigning the data points to the nearest cluster centroid, and updating the cluster centroids as the mean of the assigned data points.


````python
def kmeans(X, K):
    (n , d)=X.shape
#     M=np.zeros((K,d))
    U=np.zeros((n,K))
    M=X[np.random.choice(range(n) , size=K)]
    iteration=50
    count_iteration=0
    centroid_pre = None
    while (count_iteration<iteration and not(np.array_equal(M , centroid_pre))):
        count_iteration+=1
#         assign cluster:
        assigned=np.zeros((n,K))
#         for i in range(n):
#             distance= np.zeros(K)
#             for j in range(K):
#                 distance[j]=np.sqrt(np.sum((X[i]-M[j])**2))
#             assigned[i]=np.argmin(distance)
        for i in range(n):
            distance= np.zeros(K)
            for j in range(K):
                distance[j]=np.sqrt(np.sum((X[i]-M[j])**2))
            assigned[i][np.argmin(distance)]=1
        U=assigned
        centroid_pre=M
#         update centroid:
        M=np.zeros((K,d))
        for k in range(K):
#             finding points in each cluster: 
            cluster_points=np.empty(shape=(0, 2))
            for j in range(n):
                if (U[j][k]==1):
                    cluster_points=np.vstack((cluster_points , np.array(X[j])))
#                     cluster_points[j]=X[j]
            if cluster_points.shape[0] > 0:
                M[k]=np.mean(cluster_points , axis=0)
            
    return M, U
````

- Part 3: Write a Python function called **plot_clusters** that takes **X**, **M**, and **U** as inputs and plots the data points and the centroids on a 2-dimensional plane. Use different colors and markers for the different clusters, and draw a circle around each centroid. Label the axes as **x** and **y**, and add a title and a legend to the plot. ``(10 pt.)``

````python
def plot_clusters(X, M, U):
    K = M.shape[0]
    markers = ['1','*','.',',','+','s','D','d','p','H','h','v','^','<','>','2','3','4','o','|','_']
    colors=['c','tab:purple','tab:pink','b','g','r','y','m','tab:blue','tab:orange','tab:green','tab:red','tab:brown','tab:gray','tab:olive','tab:cyan']

    for k in range(K):
        cluster_points=np.empty(shape=(0, 2))
        for j in range(X.shape[0]):
            if (U[j][k]==1):
                cluster_points=np.vstack((cluster_points , np.array(X[j])))
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=colors[k], marker=markers[k], label=f'Cluster {k+1}')
        plt.scatter(M[k, 0], M[k, 1], c='black', marker='o')
        circle = plt.Circle((M[k, 0], M[k, 1]), radius=np.max(np.linalg.norm(cluster_points - M[k], axis=1)), facecolor='none', edgecolor='black')
        plt.gca().add_patch(circle)

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('KMeans Clustering')
    plt.legend()
    plt.show()
````

This algorithm works by minimizing the sum of squared distances between data points and their respective cluster centroids. As seen in the plot, if there is a point in more than a single cluster, we took it in the cluster which it has the least distance with it's centroid.


Advantages of using this this methods are:
1. simple: easy to understand and implement, making it accessible to a wide range of users.
2. fast and efficient: it can be efficient for large datasets, and it often converges quickly, making it suitable for real-time clustring tasks.

- Part 4: Write a Python function called **elbow_method** that takes **X** and a list of possible values for **K** as inputs and returns a numpy vector **S** of the same size as the list. **S** should contain the **sum of squared errors (SSE)** for each value of **K**, where the SSE is defined as the sum of the squared distances between each point and its assigned centroid. Plot **S** versus **K** and use the **elbow method** to find the best value of **K** for the data. Explain your reasoning and show your work. ``(15 pt.)``

````python
def elbow_method(X, K_list):
    (n,d)=X.shape
    sum_of_squared_errors = np.empty(shape=(0))
    for k in K_list:
        (M , U)= kmeans(X,k)
        sum_of_squared_distances=0
        for i in range(n):
            for j in range(k):
                sum_of_squared_distances+=(U[i][j]*np.linalg.norm((X[i]-M[j]))**2)
        sum_of_squared_errors=np.append(sum_of_squared_errors,sum_of_squared_distances)

    plt.plot(K_list, sum_of_squared_errors , marker='o')
    plt.xlabel('K')
    plt.ylabel('sum_of_squared_errors(SSE)')
    plt.title('elbow_method')
    plt.show()
    S=np.array(sum_of_squared_errors)
    return S

K_list = [i for i in range(1,11)]
S = elbow_method(X,K_list)
````

 The elbow method is a graphical method for finding the optimal K value in a k-means clustering algorithm. The elbow graph shows the sum_of_squared_errors(SSE) values on the y-axis corresponding to the different values of K (on the x-axis). The optimal K value is the point at which the graph forms an elbow or in other words, the point where the SSE decreases significantly as K increases. So based on the plot, the optimal value for K is 3.

### Preprocess the text data
In order to prepare our text data for regression analysis, we have to follow these steps:
1. Convert all text and words into lowercase
2. Removing punctuation marks such as exclamation marks, question marks, etc.
3. Creating a list of stop-words. Stop words are words that are frequently used in a language but typically do not carry significant meaning or contribute much to the sentiment of a sentence e.g. 'and', 'the', 'is', etc. **(Hint: use "re" package)**
4. Break down the text into individual words or tokens. Tokenization is a fundamental step that helps in analyzing the text at a granular level.

Comlpete the **"preprocess_text(text)"** function on the following code cell.

``(5 pt.)``

````python
def preprocess_text(text):
#     Convert all text and words into lowercase:
    text=text.lower()

#     Removing punctuation marks:
    clean_text=re.sub(r'[^\w\s]', '', text)
    
#     Creating a list of stop-words:
    stop_words=stopwords.words('english')
    
#     Breaking down the text into individual words and remove stop_words:
    words=clean_text.split()

    filtered_tokens = [word for word in words if word not in stop_words]
    text = ' '.join(filtered_tokens)
    
    return text
````

## <font color="red"> TODO : discuss BoW method </font>

The Bag-of-Words (BoW) method is a simple and popular technique for text representation in natural language processing. It involves creating a numerical representation of a text document by counting the occurrence of each word in the document. The order of words is not considered, and each word is treated as an independent entity. This method is widely used for tasks such as text classification, sentiment analysis, and information retrieval.

Here's an overview of how the Bag-of-Words method works:

1. **Tokenization**: The text is first tokenized into individual words or tokens. This process involves splitting the text into words, removing punctuation,... .

2. **Vocabulary Creation**: A vocabulary is then created by identifying all unique words in the entire corpus of documents. Each unique word becomes a feature in the vocabulary.

3. **Counting**: For each document in the corpus, a vector is created to represent the frequency of each word from the vocabulary in that document. The length of the vector is equal to the size of the vocabulary, and each element represents the count of a specific word in the document.

Python provides several built-in functions and libraries to implement the Bag-of-Words method.Here we study one of them:

 **CountVectorizer in scikit-learn**: The CountVectorizer class in the scikit-learn library provides a convenient way to implement the Bag-of-Words method. It tokenizes text, builds a vocabulary, and transforms text documents into sparse matrices representing word counts.
